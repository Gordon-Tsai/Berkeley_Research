```r
helper <- function(table){
    copy <- table$numerictime1
    table$fivemintimeStart <- copy
    table$fivemintimeEnd <- copy +300
    return(table)
}
```

```r
n11 <- helper(r11)
n12 <- helper(r12)
n13 <- helper(r13)
n15 <- helper(r15)
n16 <- helper(r16)
n17 <- helper(r17)
n1A <- helper(r1A)
n3 <- helper(r3)
n4 <- helper(r4)
n5 <- helper(r5)
n6 <- helper(r6)
n7 <- helper(r7)
n9 <- helper(r9)
nA <- helper(rA)
nB <- helper(rB)
nC <- helper(rC)
nF <- helper(rF)
```
