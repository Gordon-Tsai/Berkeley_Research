```r
add_x_degree = function(df, x){
#     """ add 10 degree to the second half of temperature"
    df$temperature[df$fivemintimeStart > mean(range(df$fivemintimeStart))] = 
                        df$temperature[df$fivemintimeStart > mean(range(df$fivemintimeStart))] + x
    return (df)
}

add_alarm = function(df, cutoff = 100){
#     """ add alarm column, True if temperature > cutoff, False otherwise """
    df$alarm = df$temperature > cutoff
    
    return (df)
}
```

```r
# randomly set temprature for each device, tempearature betwen 80 and 100
reduced_data_b3$temperature = sample(80:100, 1)
reduced_data_b4$temperature = sample(80:100, 1)
reduced_data_b5$temperature = sample(80:100, 1)
reduced_data_b7$temperature = sample(80:100, 1)
reduced_data_b9$temperature = sample(80:100, 1)

# add 10 degree to second half of temperature
reduced_data_b3 = add_x_degree(reduced_data_b3, 10)
reduced_data_b4 = add_x_degree(reduced_data_b4, 10)
reduced_data_b5 = add_x_degree(reduced_data_b5, 10)
reduced_data_b7 = add_x_degree(reduced_data_b7, 10)
reduced_data_b9 = add_x_degree(reduced_data_b9, 10)


# add alarm column, True if higher than 100, False otherwise
reduced_data_b3 = add_alarm(reduced_data_b3, 100)
reduced_data_b4 = add_alarm(reduced_data_b4, 100)
reduced_data_b5 = add_alarm(reduced_data_b5, 100)
reduced_data_b7 = add_alarm(reduced_data_b7, 100)
reduced_data_b9 = add_alarm(reduced_data_b9, 100)

layout(fmt(size(reduced_data_b3), 'b3'), fmt(size(reduced_data_b4), 'b4'), fmt(size(reduced_data_b5), 'b5'),
fmt(size(reduced_data_b7), 'b7'), fmt(size(reduced_data_b9), 'b9'))
```

```r
add_x_degree_to_pred_temp = function(df, x){
#     """ add 10 degree to the second half of temperature"
    df$pred_temp[df$fivemintimeStart > mean(range(df$fivemintimeStart))] = 
                        df$pred_temp[df$fivemintimeStart > mean(range(df$fivemintimeStart))] + x
    return (df)
}

add_pred_alarm = function(df, cutoff = 100){
#     """ add alarm column, True if temperature > cutoff, False otherwise """
    df$pred_alarm = df$pred_temp > cutoff
    
    return (df)
}
```

```r
# randomly set predicted temprature for each device, tempearature betwen 85 and 100
reduced_data_b3$pred_temp = sample(85:100, 1)
reduced_data_b4$pred_temp = sample(85:100, 1)
reduced_data_b5$pred_temp = sample(85:100, 1)
reduced_data_b7$pred_temp = sample(85:100, 1)
reduced_data_b9$pred_temp = sample(85:100, 1)

# add 10 degree to second half of pred_temp
reduced_data_b3 = add_x_degree_to_pred_temp(reduced_data_b3, 10)
reduced_data_b4 = add_x_degree_to_pred_temp(reduced_data_b4, 10)
reduced_data_b5 = add_x_degree_to_pred_temp(reduced_data_b5, 10)
reduced_data_b7 = add_x_degree_to_pred_temp(reduced_data_b7, 10)
reduced_data_b9 = add_x_degree_to_pred_temp(reduced_data_b9, 10)


# add pred_alarm column, True if pred_temp is higher than 100, False otherwise
reduced_data_b3 = add_pred_alarm(reduced_data_b3, 100)
reduced_data_b4 = add_pred_alarm(reduced_data_b4, 100)
reduced_data_b5 = add_pred_alarm(reduced_data_b5, 100)
reduced_data_b7 = add_pred_alarm(reduced_data_b7, 100)
reduced_data_b9 = add_pred_alarm(reduced_data_b9, 100)

layout(fmt(size(reduced_data_b3), 'b3'), fmt(size(reduced_data_b4), 'b4'), fmt(size(reduced_data_b5), 'b5'),
fmt(size(reduced_data_b7), 'b7'), fmt(size(reduced_data_b9), 'b9'))
```
