```r
feature <- function(array) {
    copy <- c()
    pointer <- 1
    for (i in array) {
        if (i == "False" || i == "FALSE") {
            copy[pointer] <- 0
            pointer = pointer+1
        } else {
            copy[pointer] <- 1
            pointer = pointer+1
        }
    }
    return (copy)
}
```

```r
fwd_selection <- function(table){
    pointer <- 1
    new_table <- replicate(nrow(table), 0)
    all_zero <- replicate(nrow(table), 0)
    all_one <- replicate(nrow(table),1)
    num_col <- ncol(table)
    name <- c('init')
    while (pointer < num_col){
        if (table[,pointer] == all_zero || table[,pointer] == all_one){
            pointer = pointer + 1
        } else{
            new_table <- cbind(new_table,table[,pointer])
            name <- append(name,colnames(table)[pointer])
            pointer = pointer + 1
            
        }
        new_table <- data.frame(new_table)
        colnames(new_table) <- name 
    }
    return (new_table)

}
```

```r
add_x_degree_to_pred_temp = function(df, x){
#     """ add 10 degree to the second half of temperature"
    df$pred_temp[df$fivemintimeStart > mean(range(df$fivemintimeStart))] = 
                        df$pred_temp[df$fivemintimeStart > mean(range(df$fivemintimeStart))] + x
    return (df)
}

add_pred_alarm = function(df, cutoff = 100){
#     """ add alarm column, True if temperature > cutoff, False otherwise """
    df$pred_alarm = df$pred_temp > cutoff
    
    return (df)
}
```

## Problems during data cleaning and merging
* Time column from each data in ELMS are not matched
* Time steps and columns of synthetic data and ELMS are not matched
* Insufficient data for each device and not all device have data
* Some columns are useless (all collumns are "null" or zero)

## Conclusion
* Features such as Latitude and Longitude are not useful for our predictive model (zero correlation).
* Features such as Temerature and Speed will have positive correlation, but still not sufficient to provide an accurate prediction.
* Might need other variables like distances, instrument usage (whether the AC is on or off), etc to build a more accurate model.


## Suggestions for the future
* Consider the usefulness of feature in making predictions, omit any that are not.
* Will need more data to build models that make accurate predictions.
* To build a time series model, we need a sequence of observations corresponding to adjacent time steps.
